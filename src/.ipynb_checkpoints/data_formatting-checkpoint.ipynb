# Data Formatting

```python
# Import Libraries
import pandas as pd
import ast
```

```python
# Load data
data_folder = '../data/'
df = pd.read_csv(data_folder + 'scraped_data.csv')
```

```python
# Convert a unique string in a list of strings for the following attributes
df.tags = df.tags.apply(ast.literal_eval)
df.subjects = df.subjects.apply(ast.literal_eval)
df.locations = df.locations.apply(ast.literal_eval)
df.interview_locations = df.interview_locations.apply(ast.literal_eval)
df.collections = df.collections.apply(ast.literal_eval)
```
