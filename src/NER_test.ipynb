```python
import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load('en_core_web_sm')

matcher = PhraseMatcher(nlp.vocab, attr='LOWER')

# List of names to match
names = ['John Doe', 'Alice', 'Bob']

# Convert the names to documents
patterns = [nlp(name) for name in names]

matcher.add("NamesMatcher", patterns)

# The text you want to examine
doc = nlp("John Doe and Alice are friends. Bob is Alice's brother.")

# Call the matcher on the doc
matches = matcher(doc)

# Iterate over the matches
for match_id, start, end in matches:
    # Get the matched span
    matched_span = doc[start:end]
    print(matched_span.text)
```
