```python
import pandas as pd

last_name_data = name_data["name"].apply(lambda x: x.split(". ")[-1] if ". " in x else x)
print(last_name_data)
```

```python
#name_list = ["Arthur Breslow", "C. Johnson", "Robert Ellingson"]

cleaned_names = []

for name in name_data["name"]:
    parts = name.split()

    # Remove the abbreviated part
    cleaned_parts = [part for part in parts if len(part) > 2]

    cleaned_name = " ".join(cleaned_parts)
    cleaned_names.append(cleaned_name)

print(cleaned_names)
print(len(cleaned_names))
```

```python
import re



def run(i):

  text = data.iloc[i, 1]

  Interviewee = data.iloc[i, 0]

  name_list = unique_names

  found_names = []

# Split text by paragraph
  paragraphs = re.split(r"\n\s*\n", text.strip())

# Calculate the start and end position of a paragraph
  paragraph_start_indices = [0] + [sum(len(paragraph) for paragraph in paragraphs[:i+1]) for i in range(len(paragraphs))]
  paragraph_end_indices = [sum(len(paragraph) for paragraph in paragraphs[:i+1]) for i in range(len(paragraphs))]

# Split text by sentence
  sentences = re.split(r"(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s", text.strip())

# Calculate the start and end position of a sentence
  sentence_start_indices = [0] + [sum(len(sentence) for sentence in sentences[:i+1]) for i in range(len(sentences))]
  sentence_end_indices = [sum(len(sentence) for sentence in sentences[:i+1]) for i in range(len(sentences))]

  for name in name_list:
      for paragraph_index, paragraph in enumerate(paragraphs):
          if name in paragraph:
            # Find the position of a person's name in a paragraph
              name_occurrences = [m.start() for m in re.finditer(re.escape(name), paragraph)]

            # Gets the start and end position of the paragraph
              paragraph_start = paragraph_start_indices[paragraph_index]
              paragraph_end = paragraph_end_indices[paragraph_index]

            # Find the sentence where the person's name is located
              for occurrence in name_occurrences:
                  sentence_index = None
                  for i, sentence in enumerate(sentences):
                      if paragraph_start <= sentence_start_indices[i] and sentence_end_indices[i] <= paragraph_end:
                          sentence_index = i
                          break

                # Save the person's name, location, absolute position of the paragraph, text content, paragraph number, and sentence content
                  found_names.append({
                      "Interviewee": Interviewee,
                      "name": name,
                      "start": paragraph_start + occurrence,
                      "end": paragraph_start + occurrence + len(name),
                      "paragraph_index": paragraph_index,
                      "paragraph": paragraph.strip(),
                      "absolute_start": paragraph_start + occurrence,
                      "absolute_end": paragraph_start + occurrence + len(name),
                      "paragraph_number": paragraph_index + 1,
                      "sentence": sentences[sentence_index]
                  })
  return found_names

#for name_info in found_names:
    #print(f"Name: {name_info['name']}")
    #print(f"Position: {name_info['start']} - {name_info['end']}")
    #print(f"Paragraph: {name_info['paragraph']}")
    #print(f"Absolute Position: {name_info['absolute_start']} - {name_info['absolute_end']}")
    #print(f"Paragraph Number: {name_info['paragraph_number']}")
    #print(f"Sentence: {name_info['sentence']}")
    #print()

```

```python
res = list()
for i in range(len(data)):
  nresult = run(i)
  res += nresult

df = pd.DataFrame(res, columns=['Interviewee', 'name', 'start', 'end', 'sentence', 'sentence_index', 'absolute_start', 'absolute_end'])
df
```
