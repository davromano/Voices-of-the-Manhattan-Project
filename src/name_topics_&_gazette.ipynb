```python
import pandas as pd

results_folder='../results/'

topics=pd.read_csv(results_folder + 'paragraph_topic.csv')
names = pd.read_excel(results_folder + 'Name_data.xlsx')
```

# Merge

```python
names_topics = names.merge(topics [['paragraph_index', 'topic_7','topic_30','topic_90']], on='paragraph_index', how='left') 
names_topics.head(10)
```

```python
names_topics.rename(columns={'Name_Extracted_Cleaned': 'name'}, inplace=True)
names_topics.to_excel('../results/names_topics.xlsx', index=False)
```

### Null Values

```python
# Assuming you have a DataFrame called "df"
null_values = names.isnull().sum()

print(null_values)
```

### Creating the gazette

```python
### Adding words for each topic

words=pd.read_csv('../results/lda_top20_words.csv')

words['words'] = words.apply(lambda row: row[['word_1', 'word_2', 'word_3', 'word_4', 'word_5', 'word_6', 'word_7', 'word_8', 'word_9', 'word_10', 'word_11', 'word_12', 'word_13', 'word_14', 'word_15', 'word_16', 'word_17', 'word_18', 'word_19', 'word_20']].tolist(), axis=1)
words = words.drop(columns=['word_1', 'word_2', 'word_3', 'word_4', 'word_5', 'word_6', 'word_7', 'word_8', 'word_9', 'word_10', 'word_11', 'word_12', 'word_13', 'word_14', 'word_15', 'word_16', 'word_17', 'word_18', 'word_19', 'word_20'])

# Create separate datasets based on the values in 'lda_model'
words_model_7 = words[words['model'] == 'lda_model_7']
words_model_30 = words[words['model'] == 'lda_model_30']
words_model_90 = words[words['model'] == 'lda_model_90']
words_model_7.head(7)
```

```python
# Group by 'name' column and find the most common value in 'topic_7' for each group
df_grouped = names_topics.groupby('name')['topic_7'].apply(lambda x: x.mode().iloc[0]).reset_index()
df_grouped.rename(columns={'topic_7': 'main_topic_7'}, inplace=True)

# Group by 'name' column and find the most common value in 'topic_30' for each group
df_grouped_30 = names_topics.groupby('name')['topic_30'].apply(lambda x: x.mode().iloc[0]).reset_index()
df_grouped_30.rename(columns={'topic_30': 'main_topic_30'}, inplace=True)

# Group by 'name' column and find the most common value in 'topic_90' for each group
df_grouped_90 = names_topics.groupby('name')['topic_90'].apply(lambda x: x.mode().iloc[0]).reset_index()
df_grouped_90.rename(columns={'topic_90': 'main_topic_90'}, inplace=True)

# Merge the DataFrames based on the 'name' column
results = pd.merge(df_grouped, df_grouped_30, on='name')
results = pd.merge(results, df_grouped_90, on='name')

# Print the first 10 rows of the merged DataFrame
results.head(10)

```

```python
results = pd.merge(results, words_model_7, left_on='main_topic_7', right_on='topic', how='left')
results.rename(columns={'words': 'words_topic_7'}, inplace=True)

results = pd.merge(results, words_model_30, left_on='main_topic_30', right_on='topic', how='left')
results.rename(columns={'words': 'words_topic_30'}, inplace=True)

results = pd.merge(results, words_model_90, left_on='main_topic_90', right_on='topic', how='left')
results.rename(columns={'words': 'words_topic_90'}, inplace=True)

results.drop(['topic_x', 'topic_y', 'topic'], axis=1, inplace=True)
results.drop(['model_x', 'model_y', 'model'], axis=1, inplace=True)

results.head(10)
```

```python
results.to_csv('../results/gazette.csv', index=False)
```
