# Timeline

```python
# Import Libraries
import pandas as pd
from datetime import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mpl.rcParams['figure.dpi']= 300
#plt.style.use('dark_background')
```

```python
# Load data
data_folder = '../data/scraped_data/'
df = pd.read_csv(data_folder + 'scraped_data_no_names.csv')
```

```python
df['interview_date'] = pd.to_datetime(df['interview_date'], format='%B %d, %Y', errors='coerce')
```

```python
df['interview_year'] = df['interview_date'].dt.year

count_per_year = df['interview_year'].value_counts().sort_index()

min_year = df['interview_year'].min()
max_year = df['interview_year'].max()
all_years = pd.Index(range(int(min_year), int(max_year) + 1))
count_per_year = count_per_year.reindex(all_years, fill_value=0)

plt.figure(figsize=(12,5))
count_per_year.plot(kind='bar', color='grey')
plt.xlabel("Year of the interview")
plt.ylabel("Number of interviews")
plt.box(False)
plt.grid(axis='y', color='white', linewidth=2.3)
plt.show()
```
